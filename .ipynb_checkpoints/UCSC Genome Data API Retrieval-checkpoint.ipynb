```python
from contextlib import contextmanager
import mysql.connector

@contextmanager
def connect_SQL_db(host, user):
    try:
        db = mysql.connector.connect(
                host=host,
                user=user
                )
        yield db
    except mysql.connector.Error as e:
        raise e
    else:
        db.close()
```

```python
cmd = "list/tracks?genome=" + targetGenome

# sending get request and saving the response as response object 
r = requests.get(url = URL + cmd) 

# extracting data in json format 
tracks_raw = r.json()[targetGenome]
```

```python
tracks_raw
```

```python
def parseCompositeContainer(composite, parent):
    out = {}
    for key, value in composite.items(): # if value is dict then the item is either a track or another composite
        if isinstance(value,dict):
            if "compositeContainer" in value.keys() and value["compositeContainer"]: #this module is a composit of multiple tracks
                subset = parseCompositeContainer(value, parent + " < " + key)
                out.update(subset)
                print("CONTAINER",key)
                print(len(subset.keys()))
                out[key] = {
                            "shortLabel" : value["shortLabel"],
                            "longLabel" : value["longLabel"],
                            "parent" : parent,
                            "generation" : len(parent.split(" < ")),
                            "descendents" : len(subset.keys()),
                            "track" : False
                            }

            else:
                print("-",key)
                out[key] = {
                            "shortLabel" : value["shortLabel"],
                            "longLabel" : value["longLabel"],
                            "parent" : parent,
                            "generation" : len(parent.split(" < ")),
                            "descendents" : 0,
                            "track" : True
                            }
                
    return out
```

```python
tracks = parseCompositeContainer(tracks_raw,  targetGenome)
# print(tracks["bismapBigBed"])
```

```python
GTracks = nx.Graph() 
GTracks.add_node(targetGenome)
for trackName, descr in tracks.items():
        GTracks.add_node(trackName)
        GTracks.add_edge(trackName, descr["parent"].split(" < ")[-1])

Glabels = {}
for trackName, descr in tracks.items():
    Glabels[trackName] = descr["shortLabel"]

# just because you CAN network does not mean you SHOULD
figure(num=None, figsize=(12, 12), dpi=80, facecolor='w', edgecolor='k')
nx.draw(GTracks, pos = nx.spring_layout(GTracks, k=0.1), node_size = 30, labels = Glabels)
plt.draw() 
```
